```python
from sqlalchemy import create_engine
import pandas as pd

# Replace with your own credentials
engine = create_engine('postgresql://user:password.@localhost:5432/database')

# Load the cleaned usedCars table
df = pd.read_sql("SELECT * FROM usedCars", engine)

print(df.head())
```

```python
# Price distribution
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title("Distribution of Used BMW Car Prices")
plt.show()
```

```python
# Price vs mileage
plt.figure(figsize=(10,6))
sns.scatterplot(x='mileage', y='price', data=df, alpha=0.6)
plt.title("Price vs Mileage")
plt.show()
```

```python
# Average price by model
avg_price_model = df.groupby('model')['price'].mean().sort_values(ascending=False)
avg_price_model.plot(kind='bar', figsize=(12,6))
plt.title("Average Price per BMW Model")
plt.show()
```

```python
# Price vs Engine Size
plt.figure(figsize=(10,6))
sns.scatterplot(x='enginesize', y='price', data=df, alpha=0.6)
plt.title("Price vs Engine Size")
plt.show()
```

```python
# Average price by Transmission / Fuel Type
plt.figure(figsize=(8,5))
sns.barplot(x='transmission', y='price', data=df)
plt.title("Average Price by Transmission")
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(x='fueltype', y='price', data=df)
plt.title("Average Price by Fuel Type")
plt.show()
```

```python
# Correlation heatmap
numeric_cols = df.select_dtypes(include='number')  # automatically picks ints and floats
plt.figure(figsize=(10,8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix (Numeric Features Only)")
plt.show()

```
